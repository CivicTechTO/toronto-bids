```python
import shutil
from pathlib import Path

def duplicate_folder_structure(src_folder, dst_folder):
    # Create the destination folder if it doesn't exist
    Path(dst_folder).mkdir(parents=True, exist_ok=True)
    print(f"Created destination folder: {dst_folder}")

    # Iterate through all subdirectories in the source folder
    for src_subfolder in src_folder.iterdir():
        dst_subfolder = dst_folder / src_subfolder.name
        # If the current subdirectory is a folder, recursively call the function
        if src_subfolder.is_dir():
            # Create the corresponding subdirectory in the destination folder
            dst_subfolder.mkdir(exist_ok=True)
            print(f"Created subfolder in destination: {dst_subfolder}")

            duplicate_folder_structure(src_subfolder, dst_subfolder)

def move_files(src_folder, dst_folder):
    for file in src_folder.iterdir():
        if file.is_dir():
            move_files(file, dst_folder / file)
    
    
src_folder = Path("data")
dst_folder = Path("data_processed")

duplicate_folder_structure(src_folder, dst_folder)
#move_files(src_folder, dst_folder)
```
